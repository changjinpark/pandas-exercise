```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium as f

import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'    
matplotlib.rcParams['axes.unicode_minus']=False
```

공공데이터:www.data.go.kr

```python
trafficDF = pd.read_csv('data/traffic.csv' , encoding='euc-kr')
trafficDF
```

1. 서울에서 일어난 요일별 사망자수, 사상자수 를 출력하시요

```python
g= trafficDF[ trafficDF['발생지시도'] == '서울' ].groupby('요일')
g.sum()[ ['사망자수','사상자수'] ].loc[['월','화','수','목','금','토','일']]
```

2. 서울 강북구 지역의 교통사고 위치를 지도로 표시하고 각마커에는 법규위반내용을 표시하시요

```python
kang = trafficDF[ (trafficDF['발생지시도'] == '서울') & (trafficDF['발생지시군구'] == '강북구') ][['위도','경도','법규위반']]
map = f.Map( location= ( kang['위도'].mean(), kang['경도'].mean() ),zoom_start=14 )
for i, sr in kang.iterrows():
#     print( sr['위도'],sr['경도'],sr['법규위반'])
#     mk = f.Marker( location= (sr['위도'], sr['경도']), popup=sr['법규위반'])
    test = f.Html( f"<b>{sr['법규위반']}</b>", script=True)
    htmlPop = f.Popup(test, max_width=200)
    mk = f.Marker( location= (sr['위도'], sr['경도']), popup=htmlPop)
    mk.add_to( map )  
map
```

3. 부산에서 발생한 총 사망자수를 구하시요

```python
trafficDF[ trafficDF['발생지시도'] == '부산']['사망자수'].sum()
```

4. 서울지역 월요일에 발생한 사상자수 사고유형 법규위반을 구하시요

```python
trafficDF[ ( trafficDF['발생지시도'] == '서울' ) & ( trafficDF['요일'] == '월' )][['사상자수','사고유형','법규위반']]
```

5. 횡단중 일어난 사고의 요일 발생지시도, 법규위반, 피해자_당사자종별을 구하시요

```python
trafficDF[ ( trafficDF['사고유형'] == '횡단중' ) ][['요일','발생지시도','법규위반', '피해자_당사자종별']] 
```
