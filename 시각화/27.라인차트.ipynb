```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

## matplotlib.pyplot 패키지의 plot :라인차트함수

```python
y = [10,14,19,20,25]
# plt.figure( figsize=(6,4 ) )
plt.plot(y)
plt.show()
```

```python
y = [10,14,19,20,25]
x = [1,2,3,4,5]
# plt.plot( x, y, 'o--r') # ex) r--o, r--, --
plt.plot( x, y, 's-.b') # ex) r--o, r--, --
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.plot(x,y1,'bo-',x,y2,'rs--', x,y3, 'yp-')
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
# plt.figure( figsize=(4,4 ) ) # 디폴트 figure 생략
plt.plot( x, y1, 'bo-')
plt.figure( figsize=(4,4 ) )
plt.plot( x, y2, 'rs--')
plt.plot( x, y3, 'yp-')
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.subplot(2,2,1) #행,열,인덱스
plt.plot( x,y1 )
plt.subplot(2,2,2) 
plt.plot( x,y2 )
plt.subplot(2,2,3) #행,열,인덱스
plt.plot( x,y3 )
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.subplot(1,3,1) #행,열, 넘버
plt.plot( x, y1)
plt.subplot(1,3,2)
plt.plot( x, y2)
plt.subplot(1,3,3)
plt.plot( x, y3)
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
fig, ax = plt.subplots(2,2, figsize=(7,7)) # Axes [행,열]
ax[0,0].plot( x,y1)
ax[0,1].plot( x,y2)
ax[1,0].plot( x,y3)
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
fig = plt.figure( figsize=(7,7) )
# fig.add_subplot(2,2,1)
ax1 = fig.add_subplot( 221 )
ax2 = fig.add_subplot( 222 )
ax3 = fig.add_subplot( 223 )
ax1.plot( x,y1)
ax2.plot( x,y2)
ax3.plot( x,y3)
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.xlabel( 'x축')
plt.ylabel( 'y축')
plt.title('제목')
plt.grid(True)
plt.plot(x,y1,'bo-',x,y2,'rs--', x,y3, 'yp-')
plt.legend( ['국어','영어','수학'] )
plt.show()
```

```python
x = [1,2,3,4,5]
xname = ['홍길동','이순신','임꺽정','김철수','이황']
y1 = [10,14,19,20,25]
plt.xlabel( 'x축')
plt.ylabel( 'y축')
plt.title('제목')
plt.grid(True)

plt.xticks( x, xname, rotation=45 )
plt.plot( x, y1,'ro--', alpha=0.5 )
# plt.text(1, 10, '10점')
# plt.text( 2, 14, '14점')
for a,b in zip(x,y1 ):
    plt.text( a-0.03,b+2,f'{b}점')
#     print(n)
plt.show()
```
