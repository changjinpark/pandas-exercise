```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
x=['홍길동','이순신','임꺽정','이이']
y=[90,77,50,80]
plt.bar( x,y)
plt.show()
```

```python
x=['홍길동','이순신','임꺽정','이이']
y=[90,77,50,80]
plt.barh( x,y, color=['r','g','b','y'])
plt.show()
```

### 사용할수 없는 바차트

```python
x=['홍길동','이순신','임꺽정','이이']
kor = [ 90,77,50,80 ]
eng = [60,20,70,20]

plt.bar( x,kor)
plt.bar( x,eng)
plt.legend(['국어','영어'])
plt.show() 
```

### stacked bar

```python
x=['홍길동','이순신','임꺽정','이이']
kor = [ 90,77,50,80 ]
eng = [60,20,70,20]

plt.bar( x,kor)
plt.bar( x,eng, bottom=kor )
plt.text( '홍길동',90/2, '90점',va='center',ha='center')
plt.legend(['국어','영어'])
plt.show()
```

```python
xname=['홍길동','이순신','임꺽정','이이']
x = pd.Series( [1,2,3,4] )
kor = [ 90,77,50,80 ]
eng = [60,20,70,20]
plt.xticks( x+0.2,xname)
plt.bar( x,kor,width=0.4 )
plt.bar( x+0.4, eng,width=0.4)
plt.legend(['국어','영어'])
plt.show() 
```
