```python
#키 , 몸무게
표준체중(kg)=(신장(cm)-100)×0.85
비만도(%)=현재체중/표준체중×100

비만도가
90이하:저체중, (y,100)
90∼110:정상,  (b ,150)
110∼120:과체중, (c, 200)
120이상:비만    (r ,300)
    
그래프각점에 비만도결과값(저체중, 정상,과체중, 비만)글자를
표시하고 원의 크기는  각 결과 값에 다른 크기로
그리시요.
```

```python
biman=[]
size=[]
colors=[]
def calcBiman( h,w):
    stdW = (h-100)*0.85
    s=w/stdW*100
    if s>=120: 
        size.append(300)
        biman.append('비만')
        colors.append('r')
    elif s>=110 : 
        size.append(200)
        biman.append('과제중')
        colors.append('c')
    elif s>=90 : 
        size.append(150)
        biman.append('정상')
        colors.append('b')
    else : 
        size.append(100)
        biman.append('저체중')
        colors.append('y')
```

```python
for h, w in zip( height,weight):
    calcBiman(h,w)
        
plt.scatter( height, weight,s=size, c=colors, alpha=0.5)
for h,w,b in zip(height,weight,biman):
    plt.text(h,w+2,f'{b}')
plt.show()
```
