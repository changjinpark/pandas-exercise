```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('data/chartData.csv', index_col='Unnamed: 0')
df
```

1. 국어, 영어 점수에 대한
바차트를 그리시요

```python
xname=df.index.values
x=pd.Series(range(1,11))
kor=df['국어'].values
eng=df['영어'].values
plt.xticks( x+0.2,xname)
plt.bar( x,kor,width=0.4 )
plt.bar( x+0.4, eng,width=0.4)
plt.legend(['국어','영어'])
plt.show()
```

```python
plt.bar(df.index, df['국어'])
plt.bar(df.index, df['영어'], bottom=df['국어'])
plt.legend(['국어', '영어'])
plt.show()
```

2. 영어, 수학점수에 대한
boxplot 을 그리시요

```python
plt.boxplot([df['영어'], df['수학']], labels=['영어','수학'])
plt.show()
```

3. 이름별 국어점수를 
파이차트로 그리시요

```python
plt.pie( df['국어'], labels = df.index, autopct='%.2f')
plt.show()
```

4. 국어, 영어, 수학점수를
라인차트로 그리시요

```python
# plt.plot(df.index, df['국어'],df.index,df['영어'],df.index,df['수학'] )
plt.plot(df.index, df['국어'])
plt.plot(df.index, df['영어'])
plt.plot(df.index, df['수학'])
plt.legend(['국어', '영어','수학'])
plt.show()
```
