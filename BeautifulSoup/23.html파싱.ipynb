```python
from bs4 import BeautifulSoup
import urllib.request as REQ
```

```python
kurl = 'https://news.daum.net/'
response = REQ.urlopen( kurl )
html = BeautifulSoup( response, "html5lib")
html
```

```python
ul = html.select_one( '#mArticle > div.box_headline > ul:nth-child(2)' )
for li in ul.select('li'):
#     print(li.a.string)
#     print(li.select_one('a').string)
    print(li.select_one('a strong , strong a').string)
    print("------------")
```
