#### 학년별 중간, 기말 평균을 구하시요

```python
g = df.groupby('학년')
g.mean()
```

#### 학년 과목별 중간, 기말 평균을 구하시요

```python
g = df.groupby( ['학년','과목'] )
g.mean()
```

```python
g = df.groupby( ['학년','과목'] )
g.agg( ['mean','sum'])
```

```python
g = df.pivot_table(index='학년')
g[['중간','기말']]
```

```python
g = df.pivot_table(index='학년', aggfunc='sum')
g[['중간','기말']]
```

```python
g = df.pivot_table(index='학년', aggfunc=['mean','sum'])
g
```

```python
g = df.pivot_table(index='학년', aggfunc='sum', values='중간')
g
```

```python
g = df.pivot_table(index=['학년','과목'],values='중간', columns='결과' )
g
```

```python
g = df.pivot_table(index='학년', values='중간' )
g
```
