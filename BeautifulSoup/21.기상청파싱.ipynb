```python
kdata = []
for loc in rss.select('location'):
    for data in loc.select('data'):
        kdata.append( {'도시':loc.city.string,
                      '날짜':data.tmef.string,
                      '날씨':data.wf.string,
                      '최저':int(data.tmn.string),
                      '최고':int(data.tmx.string)} )

kdata
```
